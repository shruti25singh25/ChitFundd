```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel('Exercise Data1.xlsx', index_col= 0)
```

```python
df.shape
```

```python
new_header = df.iloc[0] 
```

```python
df = df[1:]
```

```python
df.columns = new_header
```

```python
df['Net Contibution'] = df['Contribution'] - df['Amount returned to everyone in the group']
```

```python
df
```

```python
Total_net_contribution = df['Net Contibution'].sum()
Total_net_contribution
```

```python
df['Total Return'] = (df['Net amount recd by Bid winner'] - Total_net_contribution)/ Total_net_contribution
df
```

```python
df['Annualized Return'] = (((1 + df['Total Return'])**(12/25)) -1)*100
df
```

```python
last_person_bid = df.at[25,'Annualized Return']
print("Annualized Return of the person who bids in the last month is: {0} {1} ".format(last_person_bid,'%'))


#for i,j in df['Annualized Return']:
 #   print(i)
```
